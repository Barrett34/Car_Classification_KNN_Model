```python
import pandas as pd
import sklearn
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
from sklearn import linear_model, preprocessing
```

```python
data = pd.read_csv("/Users/barrett22/Desktop/Car evaluation dataset/Car Data Set/car.data")
display(data)
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
buying = label_encoder.fit_transform(list(data['buying']))
maint = label_encoder.fit_transform(list(data['maint']))
door = label_encoder.fit_transform(list(data['door']))
persons = label_encoder.fit_transform(list(data['persons']))
lugboot = label_encoder.fit_transform(list(data['lug_boot']))
safety = label_encoder.fit_transform(list(data['safety']))
_class = label_encoder.fit_transform(list(data['class']))
```

```python
predict = 'class'
```

```python
x = list(zip(buying, maint, door, persons, lugboot, safety))
y = list(_class)
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x,y, test_size=.1)
```

```python
model = KNeighborsClassifier(n_neighbors=7)
```

```python
model.fit(x_train, y_train)
```

```python
acc = model.score(x_test, y_test)
print(acc)
```

```python
prediction = model.predict(x_test)
```

```python
names = ['unacc', 'acc', 'good', 'vgood']
```

```python
for x in range(len(prediction)):
    print("predicted: ", prediction[x], "Data: ", x_test[x], "Actual: ", y_test[x])
    n = model.kneighbors([x_test[x]], 7, True)
    print("N: ", n)
```
